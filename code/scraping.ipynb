```python
import pandas as pd
from google_play_scraper import reviews, Sort
import time
pd.set_option('display.max_colwidth', None)
```

```python
app_id = 'com.spotify.music'
lang = 'en'
country = 'us'
sort = Sort.NEWEST
target_count = 500_000

all_reviews = []
continuation_token = None

while len(all_reviews) < target_count:
    print(f"Fetching... current count: {len(all_reviews)}")
    
    r, continuation_token = reviews(
        app_id,
        lang=lang,
        country=country,
        sort=sort,
        count=5000, 
        continuation_token=continuation_token
    )
    
    if not r:
        print("No more reviews to fetch.")
        break
    
    all_reviews.extend(r)

    time.sleep(1)  # delay agar tidak kena rate limit

print(f"Total collected: {len(all_reviews)} reviews")
```

```python
df = pd.DataFrame(all_reviews)
df.to_csv('../data/review_spotify.csv', index=False)
```
